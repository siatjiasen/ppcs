# Radial

```python
%pylab inline

import sigpy as sp
from demo_radial import *

import matplotlib as mpl
mpl.rcParams['text.usetex'] = True
mpl.rcParams['text.latex.preamble'] = r'\usepackage{amsmath}'
```

```python
trj = np.load("data/radial/trj.npy")
```

```python
W = sp.linop.Wavelet((256, 256))
```

```python
lp_time = np.load("results/radial/lp/time.npy")
lp_iter  = [W.H(np.load("results/radial/lp/iter_%03d.npy" % (k + 1)))[::-1, :] for k in range(lp_time.size)]
```

```python
pc_time = np.load("results/radial/pc/time.npy")
pc_iter  = [W.H(np.load("results/radial/pc/iter_%03d.npy" % (k + 1)))[::-1, :] for k in range(pc_time.size)]
```

```python
fig, axs = plt.subplots(4, 4, figsize=(10, 10), dpi=300)

def helper(axs, title, fontsize=24):
    axs.set_xticks([])
    axs.set_yticks([])
    axs.set_title(title, fontsize=fontsize)
    
lp_ref = -1
pc_ref = 3

vmin   = 0
vmax   = 0.015
errscl = 10
alpha  = 0.7

time_axis = np.linspace(max(pc_time[0], lp_time[0]), lp_time[lp_ref], 5)[:-1]

gs = axs[0, 0].get_gridspec()
for iax in axs[0:2]:
    for ax in iax[0:2]:
        ax.remove()
ax0 = fig.add_subplot(gs[0:2, 0:2])

gs = axs[0, 2].get_gridspec()
for iax in axs[0:2]:
    for ax in iax[2:4]:
        ax.remove()
ax1 = fig.add_subplot(gs[0:2, 2:4])

lst_legend = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J", "K", "L", "M", "N", "O", "P", "Q", "R"]
fontsize = 12

for k in range(4):
    
    k2 = k + 1
    k1 = np.argmin(np.abs(lp_time - pc_time[1 + k2]))
    while lp_time[k1] < pc_time[k2]:
        k1 += 1

    axs[2][k].imshow(errscl * np.abs(lp_iter[k1] - lp_iter[lp_ref]),  cmap="gray", vmin=vmin, vmax=vmax)
    axs[2][k].text(8, 47, "%s: FISTA\nWall Time: %0.3f s" % (lst_legend[k + 2], lp_time[k1]), color="white", \
                   fontsize=fontsize, bbox=dict(facecolor='black', alpha=alpha))
    
    if k < 3:
        axs[3][k].imshow(errscl * np.abs(pc_iter[k2] - lp_iter[lp_ref]), cmap="gray", vmin=vmin, vmax=vmax)
        axs[3][k].text(8, 47, "%s: Poly. Precond. FISTA\nWall Time: %0.3f s" % (lst_legend[k + 6], pc_time[k2]), color="white", \
                       fontsize=fontsize, bbox=dict(facecolor='black', alpha=alpha))
    helper(axs[2][k], r"")
    helper(axs[3][k], r"")

fig.text(0.0975, 0.314, "%d x Absolute Difference Versus (A)" % errscl, va='center', rotation='vertical', fontsize=18)


ax0.imshow(np.abs(np.squeeze(lp_iter)[lp_ref]), cmap="gray", vmin=vmin, vmax=vmax)
ax1.imshow(np.abs(np.squeeze(pc_iter)[pc_ref]), cmap="gray", vmin=vmin, vmax=vmax)

helper(ax0, "%s: FISTA\nWall Time: %0.3f s" % (lst_legend[0], lp_time[lp_ref]))
helper(ax1, "%s: Poly. Precond. FISTA\nWall Time: %0.3f s" % (lst_legend[1], pc_time[pc_ref]))

for k in range(trj.shape[0]):
    axs[-1][-1].plot(trj[k, :, 0], trj[k, :, 1])
axs[-1][-1].set_xticks([])
axs[-1][-1].set_yticks([])
axs[-1][-1].set_aspect(1)

axs[-1][-1].yaxis.set_label_position("right")
axs[-1][-1].set_ylabel(r"Trajectory", fontsize=20)

plt.subplots_adjust(wspace=-0.075, hspace=0.02)

plt.savefig('figures/radial.png')
!convert -trim figures/radial.png figures/radial.png
```
